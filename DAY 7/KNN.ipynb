```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
x = np.array([[2], [4], [6], [1]])
y = np.array(['Pass', 'Pass', 'Pass','Fail'])
```

```python
print(x)
print(y)
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(x, y)
```

```python
new_data = [[3]]
prediction = model.predict(new_data)
print("Predicted Result for 3 hours of study:", prediction[0])
```

```python
plt.figure(figsize=(8,6))

for i in range(len(x)):
    if y[i] == 'Pass':
        plt.scatter(x[i], 0, color='green', label='Pass' if i == 0 else "")
    else:
        plt.scatter(x[i], 0, color='red', label='Fail' if i == 3 else"")
        
plt.scatter(new_data, 0, color='blue', marker='x', s=100, label='New Data (3 hours)')
plt.yticks([])
plt.xlabel('Hours Studied')
plt.title('KNN Classification of Study Hours vs Exam Result')
plt.legend()
plt.grid()
plt.show()
        

    
    
```
